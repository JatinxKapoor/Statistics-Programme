```python
health = {
    'weight' : [10,20,30,40,50,60,70,80],                   #x
    'height' : [100,200,300,400,500,600,700,800]            #y
}
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(health)
```

```python
# Karl Pearson
meanx = df['weight'].sum()/len(df['weight'])
meany = df['height'].sum()/len(df['height'])
```

```python
df['x'] = df['weight']-meanx
df['y'] = df['height']-meany
```

```python
df['xy']=df['x']*df['y']
```

```python
df['x2']=df['x']**2
df['y2']=df['y']**2
```

```python
x=df['x2'].sum()
y=df['y2'].sum()
z=df['xy'].sum()
```

```python
import math
```

```python
Karl_Pearson=z/math.sqrt(x*y)
```

```python
Karl_Pearson
```

```python
#Spearman
```

```python
df['R1']=df['weight'].rank()
df['R2']=df['height'].rank()
```

```python
df['D']=df['R1']-df['R2']
```

```python
df['D2']=df['D']**2
```

```python
n=len(df)
```

```python
x=df['D2'].sum()
```

```python
spearman = 1-(6*x/n**3-n)
```

```python
spearman
```

```python
z=df['weight'].corr(df['height'],method='spearman')
z
```

```python
#outliers

iqr = df['weight'].quantile(0.75)-df['weight'].quantile(0.25)
```

```python
lx = df['weight'].quantile(0.25) - (iqr*1.5)
Ux = df['weight'].quantile(0.75) + (iqr*1.5)
```

```python
Outliers = df[(df['weight']> Ux)|(df['weight']<lx)]
print()
Outliers
```
