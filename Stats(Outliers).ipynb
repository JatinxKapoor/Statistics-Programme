```python
iqr=sub_df['Revenues'].quantile(0.75)-sub_df['Revenues'].quantile(0.25)
iqr
```

```python
lower_fence = sub_df['Revenues'].quantile(0.25) - (iqr * 1.5)
upper_fence = sub_df['Revenues'].quantile(0.75) + (iqr * 1.5)
```

```python
lower_fence
```

```python
upper_fence
```

```python
outliers = sub_df[(sub_df['Revenues'] < lower_fence) | (sub_df['Revenues'] > upper_fence)]
print('outliers rows:')
print(outliers)
```

```python
x1=sub_df['Profits'].quantile(0.75)-sub_df['Profits'].quantile(0.25)
```

```python
upper_x = sub_df['Profits'].quantile(0.75) + (1.5 * x1)
lower_x = sub_df['Profits'].quantile(0.25) - (1.5 * x1)
```

```python
x_outliers = sub_df[(sub_df['Profits'] < lower_x) |
                    (sub_df['Profits'] > upper_x)]
print('outliers rows:')
print(x_outliers)
```
