3. choose one column that has the highest correlation with the diabetes pedigree function and apply regression in order to predict the output value for a user input value.

```python
x = df['Insulin'].sum()
y = df['DiabetesPedigreeFunction'].sum()
xy = x*y
xy
```

```python
x_y_ = df['Insulin']*df['DiabetesPedigreeFunction']
xy_ = x_y_.sum()
xy_
```

```python
xx = (df["Insulin"]**2).sum()
xx
```

```python
x2 = (df["Insulin"].sum())**2
x2
```

```python
n = len(df)
n
```

```python
slope = (n*xy_ - xy)/(n*xx - x2)
slope
```

```python
intercept = (y - slope*x)/n
intercept
```
